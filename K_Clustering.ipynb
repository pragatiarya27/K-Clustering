```python
img=pydicom.dcmread("/content/swi_tra_p2_448_1800000004191561.dcm")
```

```python
img_array=img.pixel_array.astype(float)
plt.imshow(img_array,cmap="grey")
```

```python
h,w=img_array.shape
```

```python
pixels=img_array.reshape(h*w,1)
```

```python
sum_of_distance=[]
for i in range(1,11):
  model=KMeans(n_clusters=i)
  model.fit_predict(pixels)
  sum_of_distance.append(model.inertia_)
sum_of_distance
```

```python
plt.plot(range(1,11),sum_of_distance,marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Sum of Distance")
```

```python
model=KMeans(n_clusters=3)
group_number=model.fit_predict(pixels)
```

```python
segmented_images=model.cluster_centers_[group_number].reshape(h,w)
plt.imshow(segmented_images,cmap="grey")
```

```python
#Seperating the parts of the brain
for i in range(3):
  cluster_mask=group_number.reshape(h,w)==i
  segement=img_array*cluster_mask
  plt.figure()
  plt.imshow(segement,cmap="grey")

```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
```

```python
X, y =fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False) #28x28
```

```python
def plot_digits(data, title, num_rows=4, num_cols=10):
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 5))
    fig.suptitle(title, fontsize=16)
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(28, 28), cmap='gray')
        ax.axis('off')
        plt.tight_layout()
    plt.show()
```

```python
plot_digits(X[:40], "Original Digits")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
n_components = 50
pca = PCA(n_components=10)
X_reduced = pca.fit_transform(X_scaled)

X_reconstructed = pca.inverse_transform(X_reduced)
X_reconstructed = scaler.inverse_transform(X_reconstructed)

plot_digits(X_reconstructed[:40],"Reconstructed Digits")
```
